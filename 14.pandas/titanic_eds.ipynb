```python
import pandas as pd
import seaborn as sns
```

1. Загрузим датасет:

```python
data = pd.read_csv('D:/pycharm_projects/python_course/14.pandas/titanic.csv')
data.head()
```

2. Посмотрим, какая выживаемость в зависимости от пола:

```python
sns.catplot(x='Sex', y='Survived', data=data, kind='bar')
```

3. Посмотрим, какая выживаемость, в зависимости от класса:

```python
sns.catplot(x='Pclass', y='Survived', data=data, kind='bar')
```

4. Объединим эти параметры на одном графике:

```python
sns.catplot(x='Pclass', y='Survived', hue='Sex', data=data, kind='bar')
```

5. Как насчет выживаемости от количества сиблингов/мужей/жен на корабле:

```python
sns.catplot(x='SibSp', y='Survived', data=data, kind='bar')
```

6. А если от количетсва родителей/детей:

```python
sns.catplot(x='Parch', y='Survived', data=data, kind='bar')
```

7. Сравним выживаемость тех, у кого была каюта, с теми, у кого ее не было:

```python
data['has_cabin'] = data.Cabin.apply(lambda x: 0 if type(x) == float else 1)
sns.catplot(x='has_cabin', y='Survived', data=data, kind='bar')
```

8. И по возрастным группам:

```python
data['CategoricalAge'] = pd.cut(data['Age'], 5)
sns.catplot(x='CategoricalAge', y='Survived', data=data, kind='bar').set_xticklabels(rotation=30)
```

9. Блин, слишком много барплотов, надо как-то разбавить.<br>Посмотрим, как распределены люди по стоимости поездки среди мужчин и женщин:

```python
sns.set(rc={'figure.figsize': (12, 10)})

ax = sns.violinplot(x='Sex', y='Fare', data=data)
ax = sns.swarmplot(x='Sex', y='Fare', data=data, color='white', edgecolor='grey')
```
